```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import xlabel

x = ['python', 'c', 'c++', 'java', 'javascript']
y = [90, 60,70,85,80]
z = [72, 45, 60, 90, 85]
p = np.arange(len(x))
width = 0.2

plt.xlabel('programming language')
plt.ylabel('market trend')
plt.title('2025 Trend')
plt.bar(p-width/2, y, width=width, color='yellow', label='2025 popularity')
plt.bar(p+width/2, z, width=width, color='red', label='2024 popularity')
plt.xticks(p, x, rotation=20)
plt.show()
```
